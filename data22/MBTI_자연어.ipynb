```python
file = open('/content/drive/MyDrive/Colab Notebooks/MBTI프로젝트/ESTJ.txt',encoding='utf8')
ESTJ_data = file.read()
```

```python
noun_list = okt.nouns(ESTJ_data)
```

```python
wordcount = {}
# lower 소문자
for word in noun_list:  
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
```

```python
sorted(wordcount.items(), key=lambda x: x[1], reverse=True)[:10]
```

```python
file.close()
```
