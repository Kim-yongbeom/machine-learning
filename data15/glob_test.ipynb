```python
import pandas as pd
import glob
import re
from functools import reduce

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from collections import Counter

import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
import nltk
```

```python
all_files = glob.glob('./data_list/my*.xls')
all_files
```

```python
all_files_data = []
for file in all_files:
    data_frame = pd.read_excel(file)
    #print(data_frame)
    all_files_data.append(data_frame)
print(len(all_files_data))
```

```python
print(all_files_data)
```

```python
all_files_data_concat = pd.concat(all_files_data,axis=0,ignore_index=True)
all_files_data_concat
```

```python
all_title = all_files_data_concat['제목']
all_title
```

```python
# 영어 거르기
stopwords = set(stopwords.words('english'))

# 단어 추출
lemma = WordNetLemmatizer()
```

```python
words = []
for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+',' ',str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopwords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
```

```python
words
```

```python
# word가 2차원 배열이므로 1차원으로 만들어주는 작업
# reduce 반복작업해주는 함수
words2 = reduce(lambda x,y:x+y,words)
print(words2[:10])
```

```python
count = Counter(words2)
count
```

```python
word_count = dict()
count.most_common(5)
```

```python
for tag, counts in count.most_common(50):
    if len(str(tag))>1:
        word_count[tag] = counts
print(word_count)
```

```python
# 딕셔너리는 x축으로 사용불가해서 리스트로 변환
key_list = list(word_count.keys())
key_list[:10]
```

```python
sorted_keys_list = sorted(key_list)
sorted_keys_list[:10]
```

```python
sorted_values = sorted(word_count.values(), reverse=True)
sorted_values[:10]
```
