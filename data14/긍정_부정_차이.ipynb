```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file = open('neg.txt',encoding='utf8')
a = file.read()
```

```python
a
```

```python
stopwords = set(['','a','and','to','the','of','that','but','it','as','are','this','there','"','(',')','is','in','i',
                '?','-','movie','all','they','with'])
stopwords
```

```python
wordcount = {}
# lower 소문자
for word in a.lower().split():
    word = word.replace('.','')
    word = word.replace(',','')
    word = word.replace(':','')
    word = word.replace('\\','')
    word = word.replace('*','')
                        
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
```

```python
list = word_counter.most_common(5)
df1 = pd.DataFrame(list, columns=['Word','Count'])
df1
```

```python
stopwords = set(['from','about','have','had','who','was','by','with','the'
                 ,'','a','and','of','is','to','in','has','his','as','it','(',
                ')','"','that','but','film','for','he','on','this','an','are',
                'i','be',"it's",'her','all','one','then','not','when','at','they',
                ])
stopwords
```

```python
wordcount = {}
# lower 소문자
for word in b.lower().split():
    word = word.replace('.','')
    word = word.replace(',','')
    word = word.replace(':','')
    word = word.replace('\\','')
    word = word.replace('*','')
                        
    if word not in stopwords:
        if word not in wordcount:
            wordcount[word] = 1
        else:
            wordcount[word] += 1
```
